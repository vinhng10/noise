```python
import numpy as np
from scipy.signal import fftconvolve

sr = 16000

def apply_reverb(
    signal: np.ndarray, ir: np.ndarray, normalize: bool = True
) -> np.ndarray:
    # Convolve the input signal with the impulse response
    ir = ir / np.abs(ir).max()
    print(np.abs(ir).max())
    reverb_signal = fftconvolve(signal, ir, mode="full")
    reverb_signal = reverb_signal[: len(signal)]
    # reverb_signal = reverb_signal.clip(-1, 1)

    # Normalize the result if required
    if normalize:
        reverb_signal = reverb_signal / np.max(np.abs(reverb_signal))

    return reverb_signal


# Load the audio signal (e.g., "input.wav")
signal, sr = librosa.load(
    "train/data-debug/train/clean/clean_fileid_1538.wav", sr=sr, mono=True
)

# Load or generate an impulse response (e.g., "ir.wav")
ir, _ = librosa.load(
    "train/data-debug/ir/owl-ir2 (mp3cut.net).wav", sr=sr, mono=True
)

# Apply the reverb
reverb_signal = apply_reverb(signal, ir, normalize=True)
print(signal.shape, reverb_signal.shape)

display(Audio(signal, rate=sr, normalize=False))
display(Audio(reverb_signal, rate=sr, normalize=False))
```

```python
import os
import random
from pathlib import Path
import shutil

def move_random_wav_files(source_dir: str, target_dir: str, sample_fraction: float = 0.3):
    source_path = Path(source_dir)
    target_path = Path(target_dir)

    # Create target directory if it doesn't exist
    target_path.mkdir(parents=True, exist_ok=True)

    # Get all .wav files in the source directory
    wav_files = list(source_path.glob("*.wav"))

    # Randomly sample 30% of the files without replacement
    num_files_to_move = int(len(wav_files) * sample_fraction)
    files_to_move = random.sample(wav_files, num_files_to_move)

    # Move the selected files to the target directory
    for file in files_to_move:
        shutil.move(str(file), str(target_path / file.name))

# Example usage (uncomment to use):
move_random_wav_files("train/data-debug/train/ir", "train/data-debug/val/ir")
```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display


def spectrogram(path, ax, vmin=None, vmax=None):
    y, sr = librosa.load(path, sr=None)  # Load the audio file

    # Compute the Short-Time Fourier Transform (STFT) to get the spectrogram
    D = np.abs(librosa.stft(y, n_fft=2048))  # Magnitude of the STFT

    # Convert the amplitude to decibels (log scale)
    DB = librosa.amplitude_to_db(D, ref=np.max)

    # Display the spectrogram with shared vmin and vmax
    img = librosa.display.specshow(
        DB, sr=sr, x_axis="time", y_axis="log", ax=ax, vmin=vmin, vmax=vmax
    )
    ax.set(title="Spectrogram", xlabel="Time (s)", ylabel="Frequency (Hz)")

    return img  # Return the image object for colorbar use


n = 47

# Load the WAV file
noisy_path = f"train/data-10s/train-backup/noisy/noisy_fileid_{n}.wav"
clean_path = f"train/data-10s/train-backup/clean/clean_fileid_{n}.wav"

# Calculate the spectrogram for both files to determine the min and max dB values
y_clean, sr_clean = librosa.load(clean_path, sr=None)
y_noisy, sr_noisy = librosa.load(noisy_path, sr=None)

D_clean = np.abs(librosa.stft(y_clean, n_fft=512))
D_noisy = np.abs(librosa.stft(y_noisy, n_fft=512))

DB_clean = librosa.amplitude_to_db(D_clean, ref=np.max)
DB_noisy = librosa.amplitude_to_db(D_noisy, ref=np.max)

# Find global min and max dB values for consistent scaling
global_vmin = min(DB_clean.min(), DB_noisy.min())
global_vmax = max(DB_clean.max(), DB_noisy.max())

# Plot the spectrogram
fig, ax = plt.subplots(2, 1, figsize=(15, 6))
spectrogram(clean_path, ax[0], vmin=global_vmin, vmax=global_vmax)
spectrogram(noisy_path, ax[1], vmin=global_vmin, vmax=global_vmax)

# Add a colorbar for reference
# plt.colorbar(ax[0].images[0], ax=ax, orientation="vertical", format="%+2.0f dB")
plt.tight_layout()
plt.show()
```
